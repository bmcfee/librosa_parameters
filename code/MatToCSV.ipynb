```python
import numpy as np
import scipy.io
import librosa
import pandas as pd
import os
```

```python
def mat_to_csv(matfile):
    csvname = os.path.extsep.join([os.path.splitext(matfile)[0], 'csv'])
    
    data = scipy.io.loadmat(matfile)['T'].ravel()
    
    data = pd.Series(data=data)
    
    data.to_csv(csvname, index=False, mode='w')
```

```python
files = librosa.util.find_files('/home/bmcfee/data/onsets/onset_db/', ext='mat')
```

```python
for fname in files:
    print fname
    mat_to_csv(fname)
```
